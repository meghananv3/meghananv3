```
def sinplot():
  x=np.linspace(0,14,100)
  for i in range(1,5):
    plt.plot(x,np.sin(x+i*0.5)*(7-i))
```

```
sinplot()
plt.show()
```

```
sb.set_style('whitegrid')
sinplot()
plt.show()
```

```
sb.set_style("ticks")
sinplot()
plt.show()
```
